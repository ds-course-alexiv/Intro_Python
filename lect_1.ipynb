## Основные понятия

**Вычисления** — любая целенаправленная деятельность, требующая, использующая или создающая вычислительные машины, алгоритмы и программное обеспечение

**Данные** — набор дискретных или непрерывных значений, которые передают информацию, описывая количество, качество, факт, статистику, другие основные единицы смысла или последовательности символов, которые могут быть далее формально интерпретированы.
(наблюдения за явлениями реального мира.)

**Наука о данных** — междисциплинарная академическая область, которая использует статистику, вычисления, научные методы, алгоритмы для извлечения или экстраполяции знаний из больших объемов неструктурированных данных.

**Анализ данных** — процесс проверки, очистки, преобразования и моделирования данных с целью обнаружения полезной информации, информирования выводов и поддержки принятия решений.

[**Python**](https://en.wikipedia.org/wiki/Python_(programming_language)) - высокоуровневый мультипарадигмальный язык программирования общего назначения c динамической строгой типизацией и автоматическим управлением памятью, ориентированный на повышение производительности разработчика кода и удобочитаемости кода. Язык зазаботан Гвидо ван Россумом в 1989 году.

Отличительной особенностью языка является выделение блоков кода **отступами**. 

```Python
def words(document):
    collection = set()
    for line in document:
        for word in line.split():
            if word not in collection:
                collection.add(word)
    return collection
```

Язык является интерпретируемым и используется в том числе для написания скриптов; чувствителен к регистру.

**Мультипарадигмальный**, т.е поддерживает следующие элементы:
* объектно-ориентированный: любой экземпляр является объектом
* процедурный (императивный): отдельные операторы могут изменяться и объединяться в более крупные функции
* Функциональный: любой шаг является математическим вычислением функций
* Структурированный: программа представляет собой структурированную последовательность блоков кода (3 типа: последовательность или выборка, итерация).

**Высокоуровневый** — поскольку он автоматически обрабатывает самые фундаментальные операции (такие как управление памятью), выполняемые на уровне процессора ("машинный код"). Он считается языком более высокого уровня, чем, например, C, из-за его выразительного синтаксиса (который в некоторых случаях близок к естественному языку) и богатого разнообразия собственных структур данных, таких как списки, кортежи, наборы и словари.

Иногда возникает путаница в вопросе, является ли Питон языком компилируемым или интерпретируемым. Строго говоря, язык программирования сам по себе не может быть компилируемым или интерпретируемым, это свойство реализации языка. Основная реализация Питона, CPython, написанная на Си, включает в себя и компилятор, и интерпретатор. Первым делом программа на Питоне компилируется в байткод, то есть в представление на низкоуровневом языке виртуальной машины, а затем полученное представление выполняется с помощью интерпретатора байткода.

````python
import io
from graphviz import Digraph
#from railroad import Diagram, Optional, Choice, Sequence
from IPython.display import SVG, display

def show_diagram(d):
    with io.StringIO() as f:
        Diagram(d).writeSvg(f.write)
        display(SVG(f.getvalue()))
````

````python
d = Digraph()
d.attr('node', shape='box')
d.node('p', 'Программа')
d.node('c', 'Компилятор CPython')
d.node('i1', 'Интерпретатор Cpython\\n(MacOS, ARM)')
d.node('i2', 'Интерпретатор Cpython\\n(Windows, x64)')
d.node('i3', 'Интерпретатор Cpython\\n(Linux, x64)')
d.edge('p', 'c', label='Файл (*.py)')
d.edge('c', 'i1', label='Байткод (*.pyс)')
d.edge('c', 'i2', label='Байткод (*.pyс)')
d.edge('c', 'i3', label='Байткод (*.pyс)')
d
````

# Недостатки языка

1. **Нет проверок корректности до выполнения**

    Ключевым недостатком Питона можно считать практически полное отсутствие проверок корректности программы до этапа ее выполнения. Компилятор Питона своевременно оповестит программиста о синтаксических ошибках, но если вы передали в функцию, к примеру, список вместо числа, то в этом случае узнать об ошибке можно будет только во время выполнения соответствующей функции.

    В Питоне, как и в других языках с динамической типизацией, указанный недостаток до некоторой степени преодолевается с помощью написания автоматических тестов и использования сторонних инструментов статического анализа. 

2. **Скорость выполнения программ**

    Традиционно многие разработчики на Питоне жалуются на недостаточную производительность своих программ. Действительно, согласно [этим данным](https://halimshams.medium.com/python-vs-c-speed-comparison-48187511c595), на многих задачах Питон отстает от Си более чем в 100 раз. 

3. **Несовместимость со старыми версиями**

    Еще одним серьезным недостатком Питона является недостаточная совместимость между различными версиями языка 2 и 3. Они не совместимы друг с другом. Существуют проблемы со старым кодом на Питоне из учебников или из сети.

4. **Язык развивается**
 
    Следует также добавить, что многих питонистов беспокоит регулярное добавление в язык новых  конструкций, что приводит к все большему усложнению изначально простого языка.

# Среды разработки кода

## 1. Интегрированная среда разработки

**Интегрированная среда разработки** (IDE) — комплекс программных средств, используемый программистами для разработки программного обеспечения (ПО). Среда разработки включает в себя:

* текстовый редактор
* Транслятор (компилятор и/или интерпретатор)
* средства автоматизации сборки (человеческий код -> исполнимая процессором программа)

Примеры IDE:
* [VS Code](https://code.visualstudio.com/) 
* [PyCharm](https://www.jetbrains.com/pycharm/)


### VS Code

**Visual Studio Code** (VS Code) — текстовый редактор, разработанный Microsoft для Windows, Linux и macOS. Позиционируется как лёгкий и гибкий редактор кода для кроссплатформенной разработки веб- и облачных приложений. Распространяется бесплатно и включает:

* отладчик (debugger)
* подсветку синтаксиса (syntax highlighting)
* инструменты для работы с Git 
* IntelliSense = intelligent code completion
* средства для рефакторинга (повышение читаемости)
